```python
lista_empresas = df['NombreProveedor'].unique()
lista_organismos = df['NombreOrganismo'].unique()

lista_organismos[1]
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(lista_empresas, bipartite=0)
G.add_nodes_from(lista_organismos,bipartite=1)
```

```python
lista_nodos = []

for i in df['NombreOrganismo'].unique():
    for j in df[df['NombreOrganismo'] == i]['NombreProveedor'].unique():
        tupla = (i,j)
        lista_nodos.append(tupla)
```

```python
G.add_edges_from(lista_nodos)
```

```python
bipartite.is_bipartite(G)
```

```python
nx.draw_networkx(G)
```
