En el contexto de la contratación pública, los estudios anteriores se pueden dividir en dos grupos principales: (1) trabajos que exploran las relaciones bipartitas entre organismos públicos y empresas y (2) estudios que exploran las relaciones de co-licitación empresa-empresa en licitaciones públicas. ### Coordinación de Licitaciones (Empresa-Licitador?)

Para investigar más a fondo el riesgo/susceptibilidad de la manipulación del mercado por parte de las empresas, a continuación analizamos la propensión que tiene cada comunidad a participar en contratos de “postor único”. Otro patrón a menudo asociado con la corrupción y la pérdida de eficiencia. Entonces, ¿cuál es la susceptibilidad de cada comunidad a tal práctica? Para responder a esta pregunta, comenzamos investigando el número promedio de veces, por comunidad, que una empresa es el postor único de una licitación. 

Figura  5a muestra los resultados para las 22 comunidades en el componente más significativo de la red Empresa-Empresa. Tradicionalmente, un alto nivel de ofertas únicas puede ser un indicador de empresas que actúan con algún nivel de ventaja informal en el proceso de licitación o debido a la falta de competencia en un mercado específico. A nivel comunitario, dicho indicador puede ser indicativo de actividad inusual de un grupo de empresas. Por lo tanto, niveles bajos de licitación única indican el riesgo de coordinación (por ejemplo, empresas que participan coherentemente en los mismos contratos), mientras que niveles altos pueden indicar la prevalencia de mercados menos competitivos o una ventaja informal en el proceso de licitación. En general, de las diez comunidades más grandes, solo la Comunidad 8 exhibe niveles bajos de postores únicos, un patrón que se extiende también a las Comunidades 14 y 21. A diferencia de, lo que vemos en la comunidad 12 fuertemente desviada de la linea base con valor promedio de licitación única que es aproximadamente cuatro veces mayor que la de una empresa típica.

Además, analizamos el número promedio de oferentes por licitación para evaluar la posible existencia de un comportamiento coherente, es decir, la coordinación entre las empresas de una comunidad. Con ese fin, estimamos el número promedio de licitadores por licitación para cada comunidad, que normalizamos por el tamaño de la comunidad (es decir, el número de empresas en una comunidad). Curiosamente, la figura  5b muestra que en la Comunidad 8, las empresas tienden a participar en licitaciones con varias empresas que coinciden casi exactamente con el tamaño de la comunidad. Mientras que, en otros casos, como las comunidades 14 y 19, las empresas tienden a participar en licitaciones que son varias veces más grandes que sus comunidades. Vale la pena mencionar que este análisis está sesgado por el tamaño de las comunidades, por lo que la expectativa sería ver una relación que aumenta sin problemas, con la comunidad más grande logrando el valor más pequeño, y en el caso límite de una comunidad con una sola empresa tendríamos obtener el máximo. Sin embargo, es claro que en algunos casos —Comunidades 8, 14 y 19— existen aparentes desviaciones.

## Ejemplo redes Bipartitas

```python
import networkx as nx
from networkx.algorithms import bipartite
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from(['A1','A2','A3','A4'], bipartite=0)
G.add_nodes_from(['B1','B2','B3'],bipartite=1)
```

```python
G.add_edges_from([('A1', "B3"),('A4', "B1"),('A2', "B2"),('A2', "B3"),('A3', "B1")])
```

```python
bipartite.is_bipartite(G)
```

```python
nx.draw_networkx(G, pos = nx.drawing.layout.bipartite_layout(G, ['A1','A2','A3','A4']), width = 2)
```
