```python
def tb_crosstab_df(document):
   with open(document, 'rb') as f:
       result = chardet.detect(f.read())  #Detecta el tipo de caracteres utilizado en el archivo csv
   raw_data=pd.read_csv(document, encoding = result['encoding'], delimiter=','
                        , thousands='.',decimal=',',error_bad_lines=False) #Leer la data contenida en la tabla
   return raw_data
   
df = tb_crosstab_df('lic2020_hackaton.csv')
```
